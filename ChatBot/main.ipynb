```python
from langchain_core.messages import HumanMessage,AIMessage,BaseMessage
from langchain_groq import ChatGroq
from dotenv import load_dotenv
from typing import Literal,Annotated,TypedDict
from langgraph.graph import StateGraph,add_messages,END,START
```

```python
llm=ChatGroq(
    model="llama-3.3-70b-versatile"
)
```

```python
class ChatState(TypedDict):
    messages:Annotated[list[BaseMessage],add_messages]
    #add messaged is used to add messages in state 
```

```python
def chat_bot(state:ChatState):
    messages=state['messages']

    response=llm.invoke(messages)

    return  {'messages':[response]}
```

```python
graph=StateGraph(ChatState)
graph.add_node('chat_bot',chat_bot)
graph.add_edge(START,'chat_bot')
graph.add_edge('chat_bot',END)
worklflow=graph.compile()
```

```python
messages=[]
```

```python
while True:
    user_message=input("Type here :")
    print(f"User : {user_message}")

    if user_message.strip().lower() in ['exit','quit','bye']:
        break

    # Append the user's message to the history list
    messages.append(HumanMessage(content=user_message))

    # Run the workflow passing the entire conversation history list
    result = worklflow.invoke({'messages': messages})
    
    # Extract the AI's response message object
    ai_message = result['messages'][-1]
    
    # Append the AI's response to the history list
    messages.append(ai_message)

    print(f"AI : {ai_message.content}")
```
