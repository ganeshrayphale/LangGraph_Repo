```python
from langchain_groq import ChatGroq
from langgraph.graph import StateGraph ,START,END
from typing import TypedDict
from dotenv import load_dotenv

load_dotenv()
```

```python
llm=ChatGroq(
    model="llama-3.3-70b-versatile"
)

query="who is founder of Groq "
```

```python
class llmState(TypedDict):
    input_query:str
    output_ai_answer:str

```

```python
def llm_invoke(state:llmState):
    query=state['input_query']
    answer=llm.invoke(query)
    state['output_ai_answer']=answer.content

    return state
```

```python
graph=StateGraph(llmState)

graph.add_node('llm_invoke',llm_invoke)

graph.add_edge(START,'llm_invoke')
graph.add_edge('llm_invoke',END)

workflow=graph.compile()

```

```python
workflow.invoke({'input_query':query})
```
