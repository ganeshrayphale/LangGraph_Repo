```python
from langchain_groq import ChatGroq
from langchain_core.prompts import PromptTemplate
from dotenv import load_dotenv
from langgraph.graph import StateGraph,START,END
from typing import TypedDict
from langchain_core.output_parsers import SimpleJsonOutputParser

from pydantic import BaseModel,Field

import operator
from typing import Annotated
from langchain_google_genai import ChatGoogleGenerativeAI

load_dotenv()
```

```python
llm = ChatGroq(
    model="llama-3.1-8b-instant"  # Or llama-3.1-8b-instant
)
```

```python
class EssayState(TypedDict):
    essay:str
    langugae_feedback:str
    clearity_feedback:str
    analysis_feedback:str
    overall_feedback:str
    individual_score:Annotated[list[int],operator.add]
    average_score:float 
```

```python
class OuputSchema(BaseModel):
    feedback:str=Field(description="detailed feedback for the essay")
    score:int=Field(description="score out of 10",ge=0,le=10)

structured_model=llm.with_structured_output(OuputSchema)
```

```python
template=f"""evaluate the language quality of following essay genuinely and provide feedback and assign a score out of 10 \n {essay}"""
```

```python
template
```

```python
structured_model.invoke(template)
```

```python
def evaluate_language(state:EssayState):
   prompt=f"""evaluate the language quality of following essay genuinely and provide feedback and assign a score out of 10 \n {state['essay']}"""
   output=structured_model.invoke(prompt)
   return {'langugae_feedback':output.feedback,'individual_score':[output.score]}

def evaluate_analysis(state:EssayState):
   prompt=f"""evaluate the depth analysis of following essay genuinely and provide feedback and assign a score out of 10 \n {state['essay']}"""
   output=structured_model.invoke(prompt)
   return {'analysis_feedback':output.feedback,'individual_score':[output.score]}

def evaluate_clearity(state:EssayState):
   prompt=f"""evaluate the clearity in the  following essay  and genuinely  provide feedback and assign a score out of 10 \n {state['essay']}"""
   output=structured_model.invoke(prompt)
   return {'clearity_feedback':output.feedback,'individual_score':[output.score]}

def final_evaluation(state:EssayState):
    prompt=f"based on following feedback create a sumarize feedback \n language feedback {state['langugae_feedback']} \n analysis_feedback {state['analysis_feedback']} \n clearity_feedback {state['clearity_feedback']}"

    overall_feedback=llm.invoke(prompt).content
    average_score = sum(state["individual_score"]) / len(state["individual_score"])


    return  {'overall_feedback':overall_feedback,'average_score':average_score}
```

```python
graph=StateGraph(EssayState)

graph.add_node('evaluate_language',evaluate_language)
graph.add_node('evaluate_clearity',evaluate_clearity)
graph.add_node('evaluate_analysis',evaluate_analysis)
graph.add_node('final_evaluation',final_evaluation)

graph.add_edge(START,'evaluate_language')
graph.add_edge(START,'evaluate_analysis')
graph.add_edge(START,'evaluate_clearity')


graph.add_edge('evaluate_language','final_evaluation')
graph.add_edge('evaluate_analysis','final_evaluation')
graph.add_edge('evaluate_clearity','final_evaluation')

graph.add_edge('final_evaluation',END)



workflow=graph.compile()
graph.compile()
```

```python
initial_state={
    'essay': essay
}
```

```python
workflow.invoke(initial_state)
```
