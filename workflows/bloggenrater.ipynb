```python
from langchain_groq import ChatGroq
from langgraph.graph import StateGraph ,START ,END
from dotenv import load_dotenv
from typing import TypedDict
from langchain_core.prompts import PromptTemplate

load_dotenv()
```

```python
llm=ChatGroq(
    model="llama-3.3-70b-versatile"
)
```

```python
class blogState(TypedDict):
    topic:str
    outline:str
    blog:str

    
```

```python
prompt_outline=PromptTemplate(
    template="""you are given the topic {topic} create a outline on it """,
    input_variables=['topic']
)
```

```python
def get_outline(state:blogState):
    topic=state['topic']
    prompt=prompt_outline.invoke(topic)

    outline=llm.invoke(prompt).content
    state['outline']=outline
    return state
    
```

```python
prompt_blog=PromptTemplate(
    template="""you are given a topic {topic} and the {outline} prepare a professinal blog on it""",
    input_variables=['topic','outline']
)
```

```python
def get_blog(state:blogState):
    topic=state['topic']
    outline=state['outline']


    prompt=prompt_blog.invoke({'topic':topic ,'outline':outline})

    blog=llm.invoke(prompt)
    state['blog']=blog

    return state
```

```python
graph=StateGraph(blogState)

graph.add_node('outline_genrator',get_outline)
graph.add_node('blog_genrater',get_blog)

graph.add_edge(START,'outline_genrator')
graph.add_edge('outline_genrator','blog_genrater')
graph.add_edge('blog_genrater',END)

wrokflow=graph.compile()

```

```python
final_state=wrokflow.invoke({"topic":"sun"})
```
